```python
import pandas as pd
import seaborn as sns
import altair as alt
```

```python
table_explore_query = None

with open('user_table_explore.sql', 'r') as f:
    table_explore_query = f.read()

user_explore = pd.read_gbq(query=table_explore_query, project_id='lyrical-amulet-337502')

user_explore.as_of_date = user_explore.as_of_date.apply(pd.to_datetime)
```

```python
user_explore[['as_of_date', 'CNT_USERS', 'TOTAL_CHURNED', 'STARTED_DEPOSITING_TODAY', 'STARTED_WITHDRAWING_TODAY', ]].tail()
```

```python
data = user_explore[['as_of_date', 'CNT_USERS', 'TOTAL_CHURNED', 'STARTED_DEPOSITING_TODAY', 'STARTED_WITHDRAWING_TODAY']].melt(id_vars=['as_of_date'], value_vars=['CNT_USERS', 'TOTAL_CHURNED', 'STARTED_DEPOSITING_TODAY', 'STARTED_WITHDRAWING_TODAY'])
```

```python
data
```

```python
data.as_of_date = data.as_of_date.apply(pd.to_datetime)
```

```python
a = alt.Chart(data[data['variable'] != 'CNT_USERS']).mark_line().encode(
    x='as_of_date',
    y='value',
    color='variable'
) 

b = alt.Chart(data[data['variable'] == 'CNT_USERS']).mark_line().encode(
    x='as_of_date',
    y='value',
    color='variable'
) 

b | a
```
